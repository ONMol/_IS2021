### <center>Выполнение работы</center>

**Задание 1** Открыть изображение и отобразить его на экране через внешнюю программу. Также сохраните его в эту же папку под именем image1_NN (дополнительно: измените формат изображения). Выведите характеристики полученного изображения: 
- формат
- размер
- цветовой режим

**Задание 2** Поверните изображение из задания 1 image1_NN на 10-30 градусов (четные номера в журнале - по часовой стрелке, нечетные - против). Сохраните полученное изображение под именем image2_NN.

**Задание 3** Выполните обрезку изображения, оставив наиболее значимый участок изображения. Сохраните изображение под именем image3_NN.

**Задание 4** Примените к изображению из задания 3 image3_NN любой из понравившихся вам фильтров. Сохраните изображение под именем image4_NN. Примените к изображению из задания 3 image3_NN несколько (2-3) фильтров. Сохраните изображение под именем image44_NN.

**Задание 5** Подпишите изображение из задания 1 image1_NN вашей фамилией, заключив надпись в прямоугольник. При этом надпись необходимо разместить согласно номера в журнале:
- 1, 6, 11, 16, 21, 26 - в левом верхнем угле изображения;
- 2, 7, 12, 17, 22, 27 - в правом верхнем угле изображения;
- 3, 8, 13, 18, 23, 28 - в левом нижнем угле изображения;
- 4, 8, 14, 19, 24, 29 - в правом нижнем угле изображения;
- 5, 10, 15, 20, 25, 30 - в центре изображения.

Сохраните изображение под именем image5_NN.
