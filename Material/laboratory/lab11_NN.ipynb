**Тема:** Разработка и проектирование классов 

**Цель:** Получить навыки разработки, отладки и испытаний программ, содержащих описания классов и объектов.

## <center>Индивидуальные задания</center>
### <font color=blue>Задание.</font> 

Разработать класс, описывающий некоторый объект из заданной предметной области. 

Класс должен содержать: 
- строку документации; 
- не менее 6 атрибутов (переменных); 
- не менее 4 методов, выполняющих обработку атрибутов класса.

В основной программе выполнить следующее: 
- Вывести на экран документацию класса. 
- Создать и инициализировать различными значениями пять экземпляров разработанного класса.

Предметные области по вариантам:
1. Сведения об автомобилях.
2. Сведения о странах.
3. Сведения о жилых домах микрорайона.
4. Сведения о гражданах.
5. Сведения о морских судах.
6. Сведения о гражданских воздушных судах (пассажирских самолетах).
7. Сведения о военных воздушных судах.
8. Сведения о водных ресурсах.
9. Сведения о компьютерах.
10. Сведения о принтерах.
11. Сведения о сканерах.
12. Сведения о космических полетах.
13. Сведения о книгах.
14. Сведения о музыкальной коллекции.
15. Сведения о химическом элементе.
