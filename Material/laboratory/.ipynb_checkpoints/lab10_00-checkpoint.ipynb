### <center>Пример программы с использованием файлов</center>
Запишите в текстовый файл числа из списка (по одному числу в каждой строке). Создайте новый файл и перенесите в него удвоеенык значения чисел в порядке убывания из первого файла в одну текстовую строку, разделяя числа пробелами.

```python
mylist = [10, 4, 21, 17, 9, 6, 19] #Исходный список
myfile = open("files/lab10_1.txt", "w") #Открытие файла1 для записи (создание)
# Запись в файл1 чисел из списка
for x in mylist:
    print(x, file=myfile)
myfile = open("files/lab10_1.txt", "r") #Открытие файла1 для чтения
content = myfile.read() #Чтение содержимого файла1 целиком
print("Исходный файл")
print(content) #Вывод содержимого файла1 на экран
myfile.close() #Закрытие файла1
myfile = open("files/lab10_1.txt", "r") #Открытие файла1 для чтения
myfile2 = open("files/lab10_1_0.txt", "w") #Открытие файла2 для записи
mylist2=[]
#Чтение из файла1 построчно
for s in myfile:
    mylist2.append(2*int(s.rstrip('\n'))) # rstrip - функция, которая удаляет указанные символы справа
mylist2.sort() 
mylist2.reverse()
#Запись содержимого списка в файл2 с удаление служебного символа окончания строки
for s in mylist2:
    x=str(s)+" "
    myfile2.write(x)
myfile2.close() #Закрытие файла2
myfile2 = open("files/lab10_1_0.txt", "r") #Открытие файла2 для чтения
content = myfile2.read() #Чтение содержимого файла2 целиком
print("Результирующий файл")
print(content) #Вывод содержимого файла2 на экран
myfile2.close() # Закрытие файла2
    
```
