**Вариант 0.** Реализация  готовой  продукции.  Создать  родительский  класс  «Товар» (идентификатор, код, наименование, цена, описание) и дочерние классы:
- «Хрупкий товар» (коэффициент хрупкости);
- «Скоропортящийся товар» (мах время хранения);
- «Габаритный товар» (высота, ширина, длина).

Создайте 5 экземпляров различных дочерних классов (объектов) и выведите сведения о них с помощью соответствующих методов.

**Родительский класс:**

```python
class tovar:
    def __init__(self,ident,code,naim,price,description):
        self.ident = ident
        self.code = code
        self.naim = naim
        self.price = price
        self.description = description
    def show(self):
        print("Родительский объект: ",self.ident, self.code, self.naim, self.price, self.description)
```

Перекрываем родительский конструктор и дополняем его своим кодом.

**Дочерние классы:**

```python
class hrtovar(tovar):
    def __init__(self, ident,code,naim,price,description,kh):
        tovar.__init__(self, ident,code,naim,price,description)
        self.kh = kh
    def show(self):
        print("Дочерний объект: ",self.ident, self.code, self.naim, self.kh, self.price, self.description)
```

Аналогично создаем еще два дочерних класса и используем их далее.

**Создание экземпляров классов (объектов):**

```python
obj1 = tovar(1,11111,"Какой-то товар", 54.50, "LG")
obj2 = hrtovar(2,12345,"Набор бокалов", 8, 12.48, "Belarus")
```

**Вывод информации из созданных объектах объектах:**

```python
obj1.show()
obj2.show()
```
